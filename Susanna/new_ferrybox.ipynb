```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure datetime format
df_ferry['ts'] = pd.to_datetime(df_ferry['ts'])

# Set datetime as index
df_ferry.set_index('ts', inplace=True)

# Extract datetime range for titles
start_time = df_ferry.index.min().strftime('%Y-%m-%d %H:%M')
end_time = df_ferry.index.max().strftime('%Y-%m-%d %H:%M')
time_range_title = f'Data from {start_time} to {end_time}'

# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(14, 6), sharex=True)

# Plot water temperature
axes[0].plot(df_ferry.index, df_ferry['watertemp'], color='tab:blue')
axes[0].set_title(f'Water Temperature ({time_range_title})')
axes[0].set_ylabel('Â°C')
axes[0].grid(True)

# Plot salinity
axes[1].plot(df_ferry.index, df_ferry['salinity'], color='tab:green')
axes[1].set_title(f'Salinity ({time_range_title})')
axes[1].set_ylabel('PSU')
axes[1].set_xlabel('Timestamp')
axes[1].grid(True)

plt.tight_layout()
plt.show()

```
